```python
def rvrs(st):
    c = len(st)-1
    
    if c == 0:
        return st[c]
    
    return st[c] + rvrs(st[:c])
```

```python
def inverse(m):
    c = 0

    for i in range(len(m)):
        for j in range(len(m)):
            check_ = m[i][j]

            for ini in range(i, len(m)):
                for inj in range(j, len(m)):
                    with_ = m[ini][inj]
                    if check_ > with_:
                        c += 1
    return c
                    
```

```python
def inverse(m):
    
    new_m = []
    for i in m:
        new_m += i

    c = 0
    for i in range(len(new_m)):

        c2 = 0
        for j in range(i+1, len(new_m)):
            if new_m[j] < new_m[i]:
                c2 += 1

        c += c2
    
    return c

inverse(m)
```

```python
def min_max_return(arr):
    min_up = 9999999
    min_check = True
    max_up = 0

    che = 1
    for i in range(len(arr)-1):

        if arr[i] == arr[i+1]:
            che += 1
        else:
            che = 1

        if min_check:    
            if che < min_up:
                min_up = che

        if che == 1:
            min_check = False

        if che > max_up:
            max_up = che
    if max_up == len(arr):
        return 1 , 1
    return min_up, max_up

def monk_height(arr):
    arr.sort()
    if len(arr) <= 1:
        return -1
    
    max_min_up = min_max_return(arr[:len(arr)//2])
    max_min_down = min_max_return(arr[len(arr)//2:])
    
    if max(max_min_down) - min(max_min_up) == 0 and max(max_min_up) - min(max_min_down) == 0:
        return -1
        
    elif max(max_min_down) - min(max_min_up) > max(max_min_up) - min(max_min_down):
        return max(max_min_down) - min(max_min_up)
    else:
        return max(max_min_up) - min(max_min_down)
```

```python
monk_height(arr)
```

```python
arr = [1, 2, 4, 5]
print(id(arr), 'upp')
def rev(arr):
    print(id(arr), 'in func')
    
    arr[:] = arr[::-1]
    
    print(id(arr[:]), 'run')
    print(id(arr[::-1]), 'rev')
    print(id(arr))
    
print(id(arr), 'out func')
rev(arr)
print(id(arr), 'after')
```
